```python
from bank import Bank
from admin import admin_interface
from user import user_interface

def main():
    bank = Bank()
    print("""
    **********************************
    *                                *
    *      Welcome to MyBank         *
    *    Your Trusted Bank System    *
    *                                *
    **********************************\n
    """)
    while True:
        print("\n--- Welcome to MyBank ---")
        print("1. Admin Login")
        print("2. User Interface")
        print("3. Exit")
        
        try:
            choice = int(input("Enter your choice: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        
        if choice == 1:
            admin_interface(bank)
        elif choice == 2:
            user_interface(bank)
        elif choice == 3:
            print("Thank you for using MyBank. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
